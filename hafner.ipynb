```python
import json
import matplotlib.pyplot as plt

# Specify the file path
file_path = 'baseline_data/dreamer.json'

# load run data by Hafner
with open(file_path, 'r') as file:
    data = json.load(file)
```

```python
walker_walk = []
data_dict = dict()
for runs in data:
    if runs['task'] in data_dict:
        data_dict[runs['task']].append(runs)
    else:
        data_dict[runs['task']] = [runs]
```

```python
# Get the unique environments from data_dict
environments = list(data_dict.keys())

# Create a grid of plots
fig, axs = plt.subplots(4, 5, figsize=(20, 16))

# Iterate over the environments and plot the data
for i, env in enumerate(environments):
    # Calculate the row and column indices for the current environment
    row = i // 5
    col = i % 5
    
    # Get the runs for the current environment
    runs = data_dict[env]
    
    # Plot the data for each run in the current environment
    for run in runs:
        axs[row, col].plot(run['xs'][:10], run['ys'][:10], label=run['seed'])
    
    # Set the title for the current plot
    axs[row, col].set_title(env)

# Adjust the spacing between subplots
plt.tight_layout()
plt.legend()

# Show the grid of plots
plt.show()
```
